```python
import re

def preprocess_text(text):
    if not isinstance(text, str) or text.strip() == "":
        return [], ""
    
    # 1. Remove URLs
    text = re.sub(r"http\S+|www\S+", "", text)
    
    # 2. Remove emails
    text = re.sub(r"\S+@\S+", "", text)
    
    # 3. Remove numbers
    text = re.sub(r"\d+", "", text)
    
    # 4. Remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()
    
    # 5. Handle repeated characters (e.g., sooooo → soo)
    text = re.sub(r"(.)\1{2,}", r"\1\1", text)
    
    # 6. Lowercase
    text = text.lower()
    
    # 7. Tokenization
    tokens = text.split()
    
    # 8. Remove short words except "no", "not"
    cleaned_tokens = [
        word for word in tokens 
        if len(word) > 2 or word in ["no", "not"]
    ]
    
    cleaned_sentence = " ".join(cleaned_tokens)
    
    return cleaned_tokens, cleaned_sentence
```

```python
sample_inputs = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

results = []

for text in sample_inputs:
    tokens, clean_sent = preprocess_text(text)
    results.append((text, tokens, clean_sent))

# Display results
for r in results:
    print("\nOriginal:", r[0])
    print("Tokens:", r[1])
    print("Clean Sentence:", r[2])
```

```python
import numpy as np

analytics = []

for text, tokens, clean_sent in results:
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length = np.mean([len(word) for word in tokens]) if tokens else 0
    
    analytics.append((text, total_tokens, unique_tokens, avg_length))

# Display
for a in analytics:
    print("\nSentence:", a[0])
    print("Total Tokens:", a[1])
    print("Unique Tokens:", a[2])
    print("Avg Token Length:", round(a[3], 2))
```

```python
from collections import Counter

all_tokens = []

for _, tokens, _ in results:
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10 Most Frequent Words:")
print(counter.most_common(10))

print("\nTop 5 Least Frequent Words:")
print(counter.most_common()[-5:])
```

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []
    
    for text in text_list:
        tokens, clean_sent = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean_sent)
    
    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

# Test pipeline
output = full_pipeline(sample_inputs)
print(output)
```

```python
test_cases = [
    "",
    "laugh",
    "1234567890"
]

for t in test_cases:
    tokens, clean = preprocess_text(t)
    print("\nInput:", t)
    print("Tokens:", tokens)
    print("Clean Sentence:", clean)
```
